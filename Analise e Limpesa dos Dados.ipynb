```python
# 1. Importando bases
import numpy as np
import pandas as pd
dados_quest_prof         = pd.read_csv(r'./data/TS_PROFESSOR.csv', sep=',')
TS_ALUNO_9EF             = pd.read_csv(r'./data/TS_ALUNO_9EF.csv', sep=',')
```

```python
# 2. Filtros (professores)
dados_quest_prof_filtro = dados_quest_prof.query('ID_SERIE == 9') # Filtra 9° Série
dados_quest_prof_filtro = dados_quest_prof_filtro.query('ID_UF == 35') # Filtra estado de SP
dados_quest_prof_filtro = dados_quest_prof_filtro.query('IN_PREENCHIMENTO_QUESTIONARIO == 1') #Filtra ??
dados_quest_prof_filtro.query('TX_RESP_Q105 == "B" | TX_RESP_Q105 == "C"', inplace=True) # apenas professores de matemática ou ambos

# 3. Dropando vars sem sentido pós filtros
dados_quest_prof_filtro.drop(['ID_PROVA_BRASIL', 'ID_UF','ID_SERIE','IN_PREENCHIMENTO_QUESTIONARIO'], inplace=True, axis=1)

#3.1. Salvando dados que serão utilizados na tabela final (cód do professor e o questionário)
dados_tabela_final = dados_quest_prof_filtro.loc[:, 'CO_PROFESSOR':].drop_duplicates()
dados_quest_prof_filtro = dados_quest_prof_filtro.loc[:,~dados_quest_prof_filtro.columns.str.contains('^TX', case=False)] # Filtra variáveis de questões

# 4. Removendo duplicados por todas as colunas
dados_quest_prof_filtro = dados_quest_prof_filtro.drop_duplicates()

# 5. Salvando base filtrada
dados_quest_prof_filtro.to_csv(r'./data/TS_PROFESSOR_mod.csv', encoding='utf-8', index=False)
dados_tabela_final
```

```python
# 6. Filtros (alunos)
TS_ALUNO_9EF = TS_ALUNO_9EF.loc[:,~TS_ALUNO_9EF.columns.str.contains('^TX', case=False)] #Filtra questões
TS_ALUNO_9EF = TS_ALUNO_9EF.query('ID_SERIE == 9') # Filtra 9° ano
TS_ALUNO_9EF = TS_ALUNO_9EF.query('ID_UF == 35') # Filtra alunos de SP
TS_ALUNO_9EF = TS_ALUNO_9EF.query('IN_PREENCHIMENTO_PROVA == 1') # Filtra ??

# 7. Dropando vars sem sentido pós filtros
TS_ALUNO_9EF.drop(['ID_PROVA_BRASIL', 'ID_REGIAO', 'ID_UF','ID_SERIE','IN_PREENCHIMENTO_PROVA','IN_PRESENCA_PROVA'], inplace=True, axis=1)

# 8. Salvando base filtrada
TS_ALUNO_9EF.to_csv(r'./data/TS_ALUNO_9EF_mod.csv', encoding='utf-8', index=False)
```

```python
# 16. Quantidade de alunos por professor
group_data = new_df.groupby(['ID_MUNICIPIO','ID_ESCOLA','ID_DEPENDENCIA_ADM','ID_LOCALIZACAO','ID_TURMA','CO_PROFESSOR'])['ID_ALUNO'].count().reset_index().rename(columns={'ID_ALUNO':'Qtde_Aluno'})  #sum function
group_data.sort_values('Qtde_Aluno', ascending=True)

#group_data.sort_values('Qtde_Aluno', ascending=False).to_csv(r'C:\Users\wil_m\OneDrive\Área de Trabalho\TC\gruped.csv', encoding='utf-8', index=False)

#1 - Para cada professor a quantidade de alunos
#2 - A quantidade que teve nota superior a média geral do estado (vamos definir como 200).
```

```python
group_data = new_df.groupby(['ID_MUNICIPIO','ID_ESCOLA','ID_DEPENDENCIA_ADM','ID_LOCALIZACAO','ID_TURMA','CO_PROFESSOR']).agg({'ID_ALUNO': 'count','PROFICIENCIA_MT_SAEB': lambda PROFICIENCIA_MT_SAEB: (PROFICIENCIA_MT_SAEB>259).sum() }).reset_index().rename(columns={'ID_ALUNO':'Qtde_Aluno', 'PROFICIENCIA_MT_SAEB':'Qtde_Nota_Sup'})  #sum function
group_data
group_data['PercentagemAlunosAcimaDaMediaNoEstado'] = (group_data['Qtde_Nota_Sup']/group_data['Qtde_Aluno'])*100;group_data
```

```python
# 20. Criando a tabela final com o cód do professor, a resposta de suas questões e a porcentagem de alunos com nota maior que a média
merged_inner = pd.merge(left=dados_tabela_final, right=group_data,left_on='CO_PROFESSOR', right_on='CO_PROFESSOR')
merged_inner
```

```python
# 21. Dropando colunas desnecessárias 
merged_inner.drop(['ID_MUNICIPIO', 'ID_ESCOLA', 'ID_DEPENDENCIA_ADM', 'ID_LOCALIZACAO', 'ID_TURMA', 'Qtde_Aluno', 'Qtde_Nota_Sup'], inplace=True, axis=1)
merged_inner
```

```python
# 22. Mapeando valores alfabéticos em numéricos 
for col in merged_inner.columns:
    if col != 'CO_PROFESSOR' and col != 'PercentagemAlunosAcimaDaMediaNoEstado':
        merged_inner[col] = merged_inner[col].map({
                                 'A':0,
                                 'B':1,
                                 'C':2,
                                 'D':3,
                                 'E':4,
                                 'F':5,
                                 'G':6,
                                 'H':7,
                                 np.nan:10},
                                 na_action=None)
merged_inner.fillna(0, inplace=True) # removendo NaNs
merged_inner
```

```python
df = merged_inner
```

```python
merged_inner.loc[merged_inner.PercentagemAlunosAcimaDaMediaNoEstado >= 65, 'EMaiorQue65Percento'] = 1 # maior que 65% 
```

```python
merged_inner.loc[merged_inner.PercentagemAlunosAcimaDaMediaNoEstado < 65, 'EMaiorQue65Percento'] = 0 # menor que 65%
```

```python
ate65 = df.query('EMaiorQue65Percento == 0')
maior65 = df.query('EMaiorQue65Percento == 1')
maior65 # para teste de influência positiva
```

```python
merged_inner.loc[merged_inner.PercentagemAlunosAcimaDaMediaNoEstado > 35, 'EMaiorQue35Percento'] = 1 # maior que 35% 
```

```python
merged_inner.loc[merged_inner.PercentagemAlunosAcimaDaMediaNoEstado <= 35, 'EMaiorQue35Percento'] = 0 # menor que 35%
```

```python
ate35 = df.query('EMaiorQue35Percento == 0')
maior35 = df.query('EMaiorQue35Percento == 1')
ate35 # para teste de influência negativa
```
